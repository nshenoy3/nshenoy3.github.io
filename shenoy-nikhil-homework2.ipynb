## Data Visualization Homework # 2

For this assignment, I have chosen 3 photos of the sky but all of them are different in some ways. I aim to explore these differences in their RGB color space.
The photos are free to use and have been taken from https://www.freepik.com/free-photos-vectors/sky.


```python
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image as Image
```

### Image 1: 
I'd call this a 'typical' sky photo during day time - There's visibly blues and whites in here.

```python
im1 = Image.open("D:\IS445 Data Visualization\pic1.jpg")
im1
```

### Image 2:
Again, a 'typical' sky photo, but now there is some grass in this image as well. Also, visibly the blues in this photo seem to be a darker shade.

```python
im2 = Image.open("D:\IS445 Data Visualization\pic2.jpg")
im2
```

### Image 3:
This is a beautiful evening sky photo with orange, yellow, and red shades.

```python
im3 = Image.open("D:\IS445 Data Visualization\pic3.jpg")
im3
```

### Exploring the RGB color space for Image 1:

```python
im1_data = np.array(im1)
im1_data.shape
```

```python
fig = plt.figure(constrained_layout=True, figsize=(20,15))
gs = fig.add_gridspec(3, 2) # three rows, two columns

# add red to the top(0) left (0)
ax1 = fig.add_subplot(gs[0,0]) # red
ax1.hist(im1_data[:,:,0].flatten(),bins=256,color='red') # 0 = red
ax1.set_xlabel('R channel bins') 

# green will be middle (1), left (0)
ax2 = fig.add_subplot(gs[1,0]) # green
ax2.hist(im1_data[:,:,1].flatten(),bins=256,color='green') # 1 = green
ax2.set_xlabel('G channel bins') 

# blue will be bottom (2), left(0)
ax3 = fig.add_subplot(gs[2,0]) # blue
ax3.hist(im1_data[:,:,2].flatten(),bins=256,color='blue') # 2 = blue
ax3.set_xlabel('B channel bins') 

# the image will take up ALL (:) rows of the right (1) column
ax4 = fig.add_subplot(gs[:,1]) # image
ax4.imshow(im1_data)

plt.show()
```

## Writeup and Interpretation:

All the 3 images have only 3 color channels: Red, Green and Blue.
A noteworthy observation is that the scale for frequencies (Y-axis) isn't the same for all the 3 images. Hence, a quick glance at the histograms could potentially be misleading. For example, the scale for Green channel bins for Image 1 increments at intervals of 250, 500.. and so on, while the same scale for Image 2 increments at intervals of 1000.


Image 1 (clear sky) seems to have mild tones. The clouds are a shade of white which could explain why the 3 histograms have higher frequencies in the 200-255 range. Also, the image has more shades of blue towards the mid-right parts, which also could explain its Blue channel distribution.

Image 2 (clear sky with green field): The green field might explain the central nature of the green channel distribution. The shades of blue are also darker in this image, which might explain the high intensities of pixels in the 200-255 range in the blue channel.

Image 3 (evening sky): This image has darker tones, and this might explain why the histograms are skewed towards the right. The shades of orange here might explain the frequencies of the distribution for the red channel bins. There are little to none of blue shades which explains the distribution of the blue channel.

Similarities and Differences:

While all 3 images are pictures of the sky, image 1 and 2 are relatively on a lighter tones and this is evidently observed in their blue channel distributions. The presence of the green fields in the second image also explains the green channel distribution for image 2. The darker tones and orange shades in image 3 are appropriately represented in the red channel distribution. Overall, Image 3 is the one that stands apart and is different from the other two images.
